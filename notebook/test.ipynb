```python
import json
import requests
import pandas as pd

url = "https://api.thedogapi.com/v1/breeds"
resp = requests.get(url)
resp_json = json.loads(resp.text)
df = pd.DataFrame(resp_json)
df
```

```python
df.to_csv('dog_breeds_from_thedogapi.csv')
```
